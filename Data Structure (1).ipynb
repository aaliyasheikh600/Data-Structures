2.  Explain the difference between mutable and immutable data types with examples
    # Mutability & Immutability.
     - In Python, objects can be either changeable (modifiable) or unchangeable (unchangeable).
     - Mutable objects, such as lists and dictionaries, can be modified after creation, while immutable objects, such as tuples and strings, cannot.
     - Mutability affects how objects are stored and manipulated in memory, impacting performance, memory usage, and concurrency in Python programs.

4. Describe how dictionaries store data?
   - Dictionaries in Python are unordered collections that store data as key-value pairs. Each key is associated with a value, and this key-value pair is used to efficiently retrieve or manipulate data
   -Dictionaries store data as key-value pairs and use a hashing mechanism to organize the data efficiently.
   -Keys must be immutable, while values can be of any data type.
   -Dictionaries are unordered, meaning there’s no guarantee of the order in which items are stored.
   -Operations like adding, accessing, and removing key-value pairs are efficient, typically happening in constant time (O(1)) due to the hashing mechanism.

5. Why might you use a set instead of a list in Python?
   # You might use a set instead of a list in Python when:

   - Uniqueness is required – Sets automatically remove duplicate elements, whereas lists allow duplicates.
   - Fast membership testing – Checking if an item exists in a set (x in set) is faster (O(1) on average) than in a list (O(n)).
   - Set operations are needed – Sets support efficient operations like union, intersection, and difference, which are more cumbersome with lists.
  - Order is not important – Unlike lists, sets are unordered, so they are useful when order doesn’t matter.
